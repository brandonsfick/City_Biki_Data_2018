```python
import os as os 
import numpy as np
import pandas as pd
import glob
```

```python
# combined multiple csv files
path = os.getcwd() + "/Data Sets Raw" # use your path
print(path)
allFiles = glob.glob(path + "/*.csv")

list_ = []

for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)

got_data = pd.concat(list_, axis = 0, ignore_index = True)
got_data
```

```python
# Read CSV
# got_data = pd.read_csv("Data sets Cleaned/consolidated.csv")
# got_data
```

```python
# Create Trip by minutes column
got_data['Trip in Mins'] = got_data['tripduration']
got_data['Trip in Mins'] = got_data['Trip in Mins']/60
got_data
```

```python
# Drops any missing data elements
got_dataClean = got_data.dropna(how='any')
got_dataClean
```

```python
# Creates new csv
got_dataClean.to_csv('Bike_Data_Clean Last Year.csv',index=False)
```
